# Прогнозирование уровня средней заработной платы в России

```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product

warnings.filterwarnings("ignore")

def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

## Первичный анализ данных

Считываем и визуализируем данные:

```python
zp = pd.read_csv('WAG_C_M.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)

zp.head()
```

```python
plt.figure(figsize(10,5))
zp.WAG_C_M.plot()
plt.ylabel('Среднемесячная заработная плата в России')
pylab.show()
```

### STL-декомпозиция ряда

```python
plt.figure(figsize(10,7))
sm.tsa.seasonal_decompose(zp.WAG_C_M).plot()
plt.show()
```

**Краткие выводы:**

* ряд не выглядит стационарным (нужна проверка Дики-Фуллера)
* дисперсия нуждается в стабилизации, т.к. размер скачков увеличивается во времени (преобразование Бокса-Кокса)
* присутствует восходящий тренд
* наблюдается годовая сезонность

## Проверки и преобразования

Тест Дики-Фуллера:

```python
print('Dickey-Fuller test p-value =', round(sm.tsa.stattools.adfuller(zp.WAG_C_M)[1], 2))
```

Гипотеза о стационарности ряда не отвергается, беря во внимание нестабильную дисперсию применим последовательно:
    
* преобразование Бокса-Кокса
* сезонное дифференцирование
* обычное дифференцирование (на 1 лаг)

После каждого преобразования будем проверять ряд на стационарность.

```python
zp['wage_box'], lmbda = stats.boxcox(zp.WAG_C_M)
plt.figure(figsize(10, 7))
plt.plot(zp['wage_box'])
plt.ylabel('Преобразованные значения среднемесячной ЗП')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p =", round(sm.tsa.stattools.adfuller(zp['wage_box'])[1], 2))
```

```python
zp['wage_s'] = zp['wage_box'] - zp['wage_box'].shift(12)
plt.figure(figsize(10, 7))
sm.tsa.seasonal_decompose(zp.wage_s[12:]).plot()
print("Критерий Дики-Фуллера: p =", round(sm.tsa.stattools.adfuller(zp.wage_s[12:])[1], 2))
```

Хотя критерий ДФ отвергает нулевую гипотезу о стационарности ряда, из визуализации видим, что в данных сохранился тренд. Применяем последнее преобразование:

```python
zp['wage_double'] = zp['wage_s'] - zp['wage_s'].shift(1)
plt.figure(figsize(10, 7))
sm.tsa.seasonal_decompose(zp.wage_double[13:]).plot()
print("Критерий Дики-Фуллера: p =", round(sm.tsa.stattools.adfuller(zp.wage_double[13:])[1], 2))
```

## Подбор параметров и построение модели

Построим ACF и PACF полученного ряда:

```python
plt.figure(figsize(10, 7)) 
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(zp.wage_double[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(zp.wage_double[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

Начальные приближения:

```python
D = 1
d = 1

Qs = range(0, 1)
qs = range(0, 10) #установлен меньше, в целях экономии ресурсов

Ps = range(0, 1)
ps = range(0, 10) #установлен меньше, в целях экономии ресурсов
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(zp.wage_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

Лучшая модель:

```python
print(best_model.summary())
```

### Анализ остатков

```python
plt.figure(figsize(15,8))

plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Остатки **смещены** (подтверждается критерием Стьюдента), стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается критерием Льюнга-Бокса и коррелограммой).
Найдем среднюю величину остатков, чтобы скорректировать прогнозные значения:

```python
mean_resid = np.mean(best_model.resid)
```

Строим модель:

```python
zp['model'] = invboxcox(best_model.fittedvalues - mean_resid, lmbda)
plt.figure(figsize(15,7))
zp.WAG_C_M.plot()
zp.model[13:].plot(color='r')
plt.ylabel('Среднемесячная заработная плата в России')
pylab.show()
```

### Прогноз

```python
zp2 = zp[['WAG_C_M']]
date_list = [datetime.datetime.strptime("2016-09-01", "%Y-%m-%d") + relativedelta(months=x) for x in range(0,36)]
future = pd.DataFrame(index=date_list, columns= zp2.columns)
zp2 = pd.concat([zp2, future])
zp2['forecast'] = invboxcox(best_model.predict(start=len(zp[['WAG_C_M']]), end=len(zp[['WAG_C_M']])+35) - mean_resid, lmbda)

plt.figure(figsize(10,7))
zp2.WAG_C_M.plot()
zp2.forecast.plot(color='r')
plt.ylabel('Среднемесячная заработная плата в России')
pylab.show()
```
